```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize  =(5,3))
sns.boxplot(x=df['Age'], color='#FF0000')
plt.title('Bloxplot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation = numeric_df.corr()
print(correlation)
```

```python
df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'], format='%d/%m/%Y') #convert into datatime
df.info()
```

```python
#df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].astype(int)
df.info()
```

```python
grade_mapping = {'A':4.0,'B':3.0,'C':2.0,'D':1.0,'F':0.0}
df['Grade'] = df ['Grade'].map(grade_mapping)
df.info()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df['Major'] = lb_make.fit_transform(df['Major'].astype(str))
```

```python
df.drop('Name', axis=1, inplace=True)
df
```
